```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode,iplot
init_notebook_mode(connected=True)
cf.set_config_file(offline=True, world_readable=True, theme='ggplot')
from plotly import tools
import warnings
warnings.filterwarnings("ignore")
```

# 1. Producción

## 1.1 Función de producción

```python
Lm = 70 # Maximum labor
k = 10
fL = 'k*L**2' # X production function and L is labor


## End of inputs needed
L = np.arange(0,Lm+1)
y = eval(fL)
pmgL = np.diff(y)

yy = [y,pmgL]
title_large = ['Función producción X','Productividad Marginal L']
fig1 = tools.make_subplots(rows=2, cols=1, subplot_titles=(title_large))
title = ['FProd X', 'PmgL']
for i in range(2):
    trace = go.Scatter(x=L,y=yy[i],name=title[i])
    fig1.append_trace(trace, i+1,1)

fig1['layout'].update(height=600, width=700)
iplot(fig1, filename='simple-subplot')

```

## 1.2 Frontera de Posibilidades de producción

```python
L = 50 # Maximum labor
flx = '2*lx**(1/3)' # X production function
fly = '4*ly**(1/2)' # Y production function


## End of inputs needed
lx = np.arange(0,L+1)
ly = L-lx
x = eval(flx)
y = eval(fly)

# Figures code
title_large = ['Func producción Y','FPP',
          'Distribución trabajo','Func producción de X']

title = ['FProd Y','FPP',
          'Distrib L','FProd X']

xx = [ly,x,ly,lx]
yy = [y,y,lx,x]
fig = tools.make_subplots(rows=2, cols=2, subplot_titles=(title_large))
for i in range(4):
    row = [1,1,2,2]
    col = [1,2,1,2]
    trace = go.Scatter(x=xx[i],y=yy[i],name=title[i])
    fig.append_trace(trace, row[i], col[i])

fig['layout'].update(height=600, width=700)
iplot(fig, filename='simple-subplot')
```

# 2. Teoría del Consumidor

```python
# Teoría del Consumidor
RMS = '2*y/x - Px/Py' # Relación Mg de Sustitución
R = 10
Px = 2
Py = 1
#
import sympy as sym
x, y, Px, Py, R = sym.symbols('x y Px Py R')
sym.linsolve([eval(RMS), x*Px+y*Py-R], x,y)
```
